```python
from matplotlib import pyplot as plt
import numpy
```

```python
lbl = ["<50", "50-74", "75-89", "90-100"]
x = [6,14,12,1]
y = numpy.linspace(0, 6, 6)
plt.pie(x,
        labels=lbl,
        autopct="%1.1f%%", # percentage of each sector
        startangle=30, # rotate sectors
        shadow=False
    )
# plt.legend(lbl)
```

```python
w = [65,70,130,85,90,95,100,105,110,135,120,125]
plt.hist(
    w,
    [60, 70, 80, 100, 130],
    edgecolor="white"
)
plt.xlabel("weight")
plt.ylabel("count")
```

```python
matches=[1,2,3,4,5]
ascore=[3,2,1,4,5]
bscore=[1,3,2,2,4]
cscore=[2,2,3,3,2]
plt.scatter(matches,ascore,color="red",marker="+",label="ascore")
plt.scatter(matches,bscore,color="black",marker="*",label="bscore")
plt.scatter(matches,cscore,color="green",marker="x",label="cscore")
plt.legend()
```

```python
math=[78,85,90,92,95,88,84,65,74,15,76,34]
engl=[74,68,94,86,48,99,74,59,39,58,89,96]
phys=[84,22,58,75,48,78,99,95,89,73,84,93]

plt.boxplot(
    [math, engl, phys],
    labels=["math", "engl", "phys"],
    patch_artist="blue", # box fill color
    medianprops={"color":"red"}, # median line color
    boxprops={"color":"green"} # outline color
)
plt.show()
```
